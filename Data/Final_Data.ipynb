```python
import pandas as pd
import numpy as np
```

```python
df_event = pd.read_csv("Interaction_Data.csv")
df_event.head()
```

```python
df_demog = pd.read_csv("Demography_Data.csv")
df_demog.head()
```

```python
df_cat = pd.read_csv("Category_Data.csv")
df_cat.head()
```

```python
unique_visitors = df_event['visitorid'].unique()

assigned_rows = []

for vid in unique_visitors:
    np.random.seed(hash(vid) % (2**32))
    
    rand_idx = np.random.choice(df_demog.index)
    
    assigned_rows.append(df_demog.loc[rand_idx])

df_user_map = pd.DataFrame({'visitorid': unique_visitors}).reset_index(drop=True)
df_user_map = pd.concat([df_user_map, pd.DataFrame(assigned_rows).reset_index(drop=True)], axis=1)

df_user_map.head()
```

```python
unique_items = df_event['itemid'].unique()

assigned_categories = []

for iid in unique_items:
    np.random.seed(hash(iid) % (2**32))
    
    rand_idx = np.random.choice(df_cat.index)
    
    assigned_categories.append(df_cat.loc[rand_idx,'category'])
    
df_item_map = pd.DataFrame({
    'itemid': unique_items,
    'category': assigned_categories
})

df_item_map.head(10)
```

```python
df_merged = pd.merge(df_event, df_user_map, on='visitorid', how='left')

df_final = pd.merge(df_merged, df_item_map, on='itemid', how='left')

df_final.head(10)
```

```python
df_final.to_csv('Final_Data.csv', index=False)
```
