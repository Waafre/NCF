```python
duplikat_detail = (
    df_event.groupby(['visitorid', 'itemid', 'event', 'timestamp'])
      .size()
      .reset_index(name='count')
      .query('count > 1')
      .sort_values('count', ascending=False)
)

duplikat_detail.head(10)
```

```python
#delete
df_event = df_event.drop_duplicates(subset=['visitorid', 'itemid', 'event', 'timestamp'])

#cek
df_event.duplicated(subset=['visitorid', 'itemid', 'event', 'timestamp']).sum()
```

```python
jumlah_user_unik = df_event['visitorid'].nunique()
jumlah_item_unik = df_event['itemid'].nunique()

print(f"Jumlah Pengguna Unik (visitorid): {jumlah_user_unik:,}")
print(f"Jumlah Item Unik (itemid): {jumlah_item_unik:,}")
```

```python
jumlah_event = df_event['event'].value_counts()
print(jumlah_event)

persentase_event = df_event['event'].value_counts(normalize=True) * 100
print(f"\n{persentase_event.round(2).astype(str) + ' %'}")
```

```python
score_mapping = {
    'view': 1,
    'addtocart': 2,
    'transaction': 3
}

df_event = df_event.copy()
df_event['rating'] = df_event['event'].map(score_mapping)

df_event.head()
```
